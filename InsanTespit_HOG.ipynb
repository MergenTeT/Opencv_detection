```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

HOG yöntemi uygulanıyor

```python
hog = cv.HOGDescriptor()
hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())
```

```python
img = cv.imread("x720.jpg")
```

```python
img = imutils.resize(img,width=min(400,img.shape[1]))

regions,_ = hog.detectMultiScale(img,padding = (4,4),scale = 1.05)
```

```python
for (x, y, w, h) in regions:
    cv.rectangle(img, (x, y), 
                  (x + w, y + h), 
                  (0, 0, 255), 2)
```

```python
plt.imshow(img)
```
